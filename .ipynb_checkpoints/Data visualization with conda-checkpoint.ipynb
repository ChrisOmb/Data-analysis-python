```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    data = pd.read_csv('countries.csv')
    print("Data loaded successfully!")
except FileNotFoundError:
    print("Error: File not found.")
    exit()
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

```python
if data.isnull().sum().any():
    print("Missing values detected. Handling them...")
    data = data.dropna()
```

```python
print(f"Unique countries: {data['country'].nunique()}")
most_populous = data.loc[data['population'].idxmax()]
print(f"Most populous country and year: {most_populous}")
```

```python
country = 'India'
country_data = data[data['country'] == country]
plt.figure(figsize=(10, 6))
plt.plot(country_data['year'], country_data['population'], marker='o', label=country)
plt.title(f'Population Trend of {country}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```
